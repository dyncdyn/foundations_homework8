```python
from selenium import webdriver
import requests
import re
import pandas as pd
```

```python
url = 'http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf'
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
heads = []
headlines = driver.find_elements_by_class_name("titlebet")
for head in headlines:
    heads.append(head.text)
len(heads)
```

```python
list_ = []
for head in headlines:
    onclick = head.get_attribute('onclick')
    list_.append(onclick)
len(list_)
```

```python
IDs = []
for li in list_:
    ID = re.findall('AR\d{7}',li)
    IDs.append(ID[0])
len(IDs)
```

```python
nk_news = pd.DataFrame(
    {'head': heads,
     'onclick':list_,
     'id':IDs
    }
)
nk_news
```

```python
nk_news.to_csv('NKnews.csv',index=False)
```
